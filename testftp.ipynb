```python
import socket
from itertools import count


def send_tcp_file(target_ip, target_port, file_path):
    with open(file_path, 'rb') as file:
        data = file.read()

    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((target_ip, target_port))
        s.sendall(data)

        # Optionally receive response
        response = s.recv(1024)
        print("Received response:", response.decode('utf-8'))

# Example usage
target_ip = "192.168.2.101"
target_port = 21
file_path = "data.txt"

for i in count (0):
    send_tcp_file(target_ip, target_port, file_path)
```

```python
import socket
import struct

def send_raw_tcp_frame(destination_ip, destination_port, data):
    # Create a raw socket
    raw_socket = socket.socket(socket.AF_INET, socket.SOCK_RAW, socket.IPPROTO_RAW)

    # Set the IP header manually (you might need to adjust this based on your needs)
    ip_header = struct.pack('!BBHHHBBH4s4s', 0x45, 0, 20 + len(data), 0, 0, 6, 0, 0, socket.inet_aton('192.168.1.1'), socket.inet_aton(destination_ip))

    # Combine the IP header and the data
    packet = ip_header + data

    # Send the packet
    raw_socket.sendto(packet, (destination_ip, destination_port))

# Example usage
destination_ip = '192.168.2.101'
destination_port = 21
data = b'Hello, this is a raw TCP frame!'

send_raw_tcp_frame(destination_ip, destination_port, data)
```
